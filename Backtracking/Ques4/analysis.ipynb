```python
def fair_partition(arr, k):
    def is_feasible(mid):
        count = 1
        current_sum = 0
        partitioned_arr = [[]]

        for num in arr:
            if current_sum + num > mid:
                count += 1
                current_sum = 0
                partitioned_arr.append([])

            current_sum += num
            partitioned_arr[-1].append(num)

        return count <= k, partitioned_arr

    low, high = max(arr), sum(arr)

    while low <= high:
        mid = (low + high) // 2

        feasible, partitioned_arr = is_feasible(mid)

        if feasible:
            high = mid - 1
        else:
            low = mid + 1

    return partitioned_arr



```

```python
arr = [100, 400, 300, 400, 500, 600, 700, 800, 100]
k = 3 #no of guys to be partitioned
result = fair_partition(arr, k)
print("Fair partitioning", result)
```

```python
import time
import random
time_array = list()

for i in range(9, 15,2):
    arr = [random.randint(100, 500) for _ in range(i)]
    start_time = time.perf_counter()
    print(f"For N = {i}:\n{fair_partition(arr, 3)}\n")
    end_time = time.perf_counter()
    time_array.append(end_time - start_time)
print(time_array)
```

```python
import matplotlib.pyplot as plt

n = [9,11,13]
plt.plot(n,time_array, color='r', label='Execution Time')
plt.ylabel('Execution Time (seconds)')
plt.title('Input Size VS Execution Time')
plt.grid(True)
plt.legend()
plt.show()
```
