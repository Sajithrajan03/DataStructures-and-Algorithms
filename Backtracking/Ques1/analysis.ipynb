```python
def is_safe(person, invited, enemies):
    return not any(enemy in invited for enemy in enemies[person])

def invite_friends(friends, enemies):
    sorted_friends = sorted(friends, key=lambda x: len(enemies[x]), reverse=True)

    invited = set()

    for person in sorted_friends:
        if is_safe(person, invited, enemies):
            invited.add(person)

    return invited

```

```python
import time
time_array = list()

friends = ['Sajith', 'Ananya', 'Praveen', 'Karthick']
enemies = {
    'Sajith': {'Ananya', 'Praveen'},
    'Ananya': {'Sajith', 'Karthick'},
    'Praveen': {'Sajith', 'Karthick'},
    'Karthick': {'Ananya', 'Praveen'}
}

friends_1 = ['Sajith', 'Ananya', 'Praveen', 'Karthick']
enemies_1 = {
    'Sajith': {'Ananya'},
    'Ananya': {'Sajith'},
    'Praveen': {'Karthick'},
    'Karthick': {'Praveen'}
}

  
friends_2 = ['Sajith', 'Ananya', 'Praveen', 'Karthick']
enemies_2 = {
    'Sajith': {'Ananya'},
    'Ananya': {'Sajith', 'Karthick'},
    'Praveen': {'Karthick'},
    'Karthick': {'Praveen'}
}




start_time = time.perf_counter()
print(f"\n{enemies}\n{invite_friends(friends, enemies)}\n")
end_time = time.perf_counter()
time_array.append(end_time - start_time)

start_time = time.perf_counter()
print(f"\n{enemies_1}\n{invite_friends(friends_1, enemies_1)}\n")
end_time = time.perf_counter()
time_array.append(end_time - start_time)

start_time = time.perf_counter()
print(f"\n{enemies_2}\n{invite_friends(friends_2, enemies_2)}\n")
end_time = time.perf_counter()
time_array.append(end_time - start_time)

print(time_array)
```

```python
import matplotlib.pyplot as plt

n = [1,2,3]
plt.plot(n,time_array[::-1], color='r', label='Execution Time')
plt.ylabel('Execution Time (seconds)')
plt.title('Input Size VS Execution Time')
plt.grid(True)
plt.legend()
plt.show()
```
